```python
stop=stopwords.words('english')
```

```python
import string
pun=list(string.punctuation)
```

```python
stop=stop+pun
```

```python
text="hello there my name is shailesh iam learning machine learning from coding ninjas and so far it been a wonderful journy"
```

```python
documents=[[word_tokenize(df1.iloc[i,1]),df1.iloc[i,0]] for i in range(df1.shape[0])]
```

```python
documents[0:5]
```

```python
len(documents)
```

```python
import random
random.seed(0)
random.shuffle(documents)
```

```python
training_document=documents[500:]
testing_document=documents[:500]
```

```python
all_words=[]
for doc in training_document:
    for w in doc[0]:
        if w.lower() not in stop:
            all_words.append(w.lower())
```

```python
len(all_words)
```

```python
word_dis=nltk.FreqDist(all_words)
```

```python
feature=word_dis.most_common(3000)
```

```python
feature_words=[i[0] for i in feature]
```

```python
feature_words
```

```python
def get_feature(document):
    word=set(document)
    feature={}
    for w in feature_words:
        feature[w]=(w in word)
    return feature
```

```python
training_data=[(get_feature(a),b) for (a,b) in training_document]
testing_data=[(get_feature(a),b) for (a,b) in testing_document]
```

```python
classifier = nltk.NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier, testing_data)
```

```python
import pickle 
```

```python
classifier_file = open("twitter.pickle","wb")
pickle.dump(classifier,classifier_file)
classifier_file.close()
```

```python
classifier_new = open("twitter.pickle","rb")
classifier_copy = pickle.load(classifier_new)
classifier_new.close()
```
